```python
import geocoder
import pandas
import folium
```

```python
data = pandas.read_excel('Air Pollution Monitoring Highbury East_10_01_17.xlsx')
```

```python
def find_coords(address):
    g = geocoder.google(address+',UK')
    return g.latlng

data['coords'] = data.Street.map(find_coords)
```

```python
center = [51.5551275, -0.09601359999999999]
zoom = 15

map_2011 = folium.Map(location = center, zoom_start = zoom)
map_2016 = folium.Map(location = center, zoom_start = zoom)
colours_map = {'NO² (60-80) 50% Above EU Limits':'red', 'NO² (<40) Below EU Limits':'blue',
       'NO² (40-60) Above EU Limits':'yellow'}

for _,row in data.iterrows():
    if row['Sept 2011 results'] == row['Sept 2011 results']:
        folium.RegularPolygonMarker(location=row['coords'], popup=row['Street'],
                   fill_color=colours_map[row['Sept 2011 results']], number_of_sides=8, radius=8).add_to(map_2011)
    if row['Nov 2016 results.1'] == row['Nov 2016 results.1']:
        folium.RegularPolygonMarker(location=row['coords'], popup=row['Street'],
                    fill_color=colours_map[row['Nov 2016 results.1']], number_of_sides=8, radius=8).add_to(map_2016)
    
```
